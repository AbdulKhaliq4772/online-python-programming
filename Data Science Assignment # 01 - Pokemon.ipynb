```python
HP_mean=data["HP"].mean()
def set_HP(HP_value):
    if HP_value<HP_mean:
        return "HP low"
    else:
        return "HP High"
data["HP High Low"]=data["HP"].apply(set_HP)

speed_mean=data["Speed"].mean()
def set_speed(speed_value):
    if speed_value<speed_mean:
        return "Speed Low"
    else:
        return "Speed High"
data["Speed High Low"]=data["Speed"].apply(set_speed)
data
```
